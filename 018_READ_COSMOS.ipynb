```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
zCOSMOS = pd.read_csv("../data/cosmos/cesam_zcosbrightspec20k_dr3_catalog_1559644675.csv")
```

```python
zCOSMOS.columns.values
```

```python
plt.figure(figsize=(10,8))
plt.plot(zCOSMOS['ra'][zCOSMOS['ra']>0],zCOSMOS['dec'][zCOSMOS['ra']>0],",")
plt.xlabel('RA',fontsize=30)
plt.ylabel('DEC',fontsize=30)
plt.tick_params(labelsize=15)
plt.title('zCOSMOS bright 20k DR3 z-spec',fontsize=30)
plt.ylim(1.4,3.1)
plt.xlim(149.25,151.1)
plt.savefig('../plots/018/zCOSMOS_radec.png')
plt.show()
```

```python
#fits_image_filename = fits.util.get_testdata_filepath('COSMOS2015_Laigle_v1.1.fits')
hdul = fits.open('../data/cosmos/COSMOS2015_Laigle_v1.1.fits')
cat = hdul[1].data
print(cat.names) # print the names of the columns
# dir(cat) # it says which commands I can use
```

```python
plt.figure(figsize=(10,8))
plt.plot(cat["ALPHA_J2000"],cat["DELTA_J2000"],",")
plt.xlabel("RA [J2000]",fontsize=30)
plt.ylabel("DEC [J2000]",fontsize=30)
plt.tick_params(labelsize=15)
plt.title("COSMOS Laigle+15 (photo-z)",fontsize=30)
plt.ylim(1.4,3.1)
plt.xlim(149.25,151.1)
plt.savefig('../plots/018/COSMOS_Laigle15_radec.png')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(cat["ALPHA_J2000"][cat['B_MAG_APER2']<0],cat["DELTA_J2000"][cat['B_MAG_APER2']<0],",",label='B_MAG_APER2 < 0')
#plt.SCATTER(cat["ALPHA_J2000"],cat["DELTA_J2000"])
plt.xlabel("RA [J2000]",fontsize=30)
plt.ylabel("DEC [J2000]",fontsize=30)
plt.tick_params(labelsize=15)
plt.title("COSMOS Laigle+15 (photo-z)",fontsize=30)
plt.ylim(1.4,3.1)
plt.xlim(149.25,151.1)
plt.legend(loc=1)
plt.savefig('../plots/018/COSMOS_Laigle15_bad_values_B.png')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(cat["ALPHA_J2000"][cat['H_MAG_APER2']<0],cat["DELTA_J2000"][cat['H_MAG_APER2']<0],",",label='H_MAG_APER2 < 0')
#plt.SCATTER(cat["ALPHA_J2000"],cat["DELTA_J2000"])
plt.xlabel("RA [J2000]",fontsize=30)
plt.ylabel("DEC [J2000]",fontsize=30)
plt.tick_params(labelsize=15)
plt.title("COSMOS Laigle+15 (photo-z)",fontsize=30)
plt.ylim(1.4,3.1)
plt.xlim(149.25,151.1)
plt.legend(loc=1)
plt.savefig('../plots/018/COSMOS_Laigle15_bad_values_H.png')
plt.show()
```
