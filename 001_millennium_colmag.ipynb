```python
v1,u1,z1,mass1 = np.loadtxt("data_random00009.csv",unpack=True,delimiter=",",usecols=(0,1,2,4))
#v1,u1,z1,mass1 = np.loadtxt("data_z0413_random00009.csv",unpack=True,delimiter=",",usecols=(0,1,2,4))
```

```python
print len(v1),"galaxies from Gonzalez2014a..MR7"
```

```python
plt.plot(v1,u1-v1,",")
plt.xlabel("mag V ext")
plt.ylabel("mag U - mag V ext")
plt.ylim(-0.5,2.5)
plt.title("Total colour-magnitude for "+str(len(v1))+" galaxies")
plt.show()
```

```python
h=0.73
log_mass = np.log10(mass1*10**10/h) 
high_mass = np.where(log_mass>8.5)
```

```python
plt.hist(log_mass,bins=40)
plt.xlabel(r"log Stellar mass $[M_{\odot}]$")
plt.ylabel("counts")
plt.title("distribution of stellar masses for the total sample")
plt.show()
```

```python
plt.plot(v1[high_mass],u1[high_mass]-v1[high_mass],",")
plt.xlabel("mag V ext")
plt.ylabel("(mag U - mag V) ext")
#plt.ylim(-1,2.5)
plt.show()
```

```python
zint1 = np.where((z1>0.4)&(z1<=0.5))
zint2 = np.where((z1>0.5)&(z1<=0.6))
zint3 = np.where((z1>0.6)&(z1<=0.7))
zint4 = np.where((z1>0.7)&(z1<=0.8))
zint5 = np.where((z1>0.8)&(z1<=0.9))
zint6 = np.where((z1>0.9)&(z1<=1.0))
zint7 = np.where((z1>1.0)&(z1<=1.3))
sel = [zint1,zint2,zint3,zint4,zint5,zint6,zint7]
sel = sel[::-1]
```

```python
plt.plot(v1[zint1],u1[zint1]-v1[zint1],",")
plt.plot(v1[zint2],u1[zint2]-v1[zint2],",")
plt.plot(v1[zint3],u1[zint3]-v1[zint3],",")
plt.plot(v1[zint4],u1[zint4]-v1[zint4],",")
plt.plot(v1[zint5],u1[zint5]-v1[zint5],",")
plt.plot(v1[zint6],u1[zint6]-v1[zint6],",")
plt.plot(v1[zint7],u1[zint7]-v1[zint7],",")
plt.show()
```

```python
style = ["k,","y,","m,","c,","r,","g,","b,"]
zz = [0.4,0.5,0.6,0.7,0.8,0.9,1.0,1.3]
style = style[::-1]
fig = plt.figure(figsize=(10,15))

#str(zz[i])+" < z < "+str(zz[i+1])

for i in (range(len(sel))):
    select = np.where((z1>zz[i])&(z1<=zz[i+1]))

    ax = fig.add_subplot(4,2,7-i)
    ax_last = fig.add_subplot(4,2,8)
    ax_last.set_xlabel("mag V ext")
    ax_last.plot(v1[select],u1[select]-v1[select],style[i])
    ax.plot(v1[select],u1[select]-v1[select],style[i],label=str(zz[i])+" < z < "+str(zz[i+1]))
    ax.legend(numpoints=1)
    ax_last.set_xlim(-24.4,-8.1)
    ax_last.set_ylim(-0.25,2.45)
    ax_last.axvline(-22,linestyle='dashed')
    
    ax.set_xlim(-24.4,-8.1)
    ax.set_ylim(-0.25,2.45)
    ax.axvline(-22,linestyle='dashed')
    
    if (7-i)!=len(sel) and (7-i)!= len(sel)+1: 
        plt.setp(ax.get_xticklabels(),visible=False,fontsize=8)
    else:
        plt.setp(ax.get_xticklabels(),visible=True,fontsize=8)
        ax.set_xlabel("mag V ext")
    
    if np.mod((i+1),2)==0:
        plt.setp(ax.get_yticklabels(),visible=False,fontsize=8)
    else:
        plt.setp(ax.get_yticklabels(),visible=True,fontsize=8)
        ax.set_ylabel("(mag U - mag V) ext") 

plt.subplots_adjust(wspace=0,hspace=0)


plt.legend(["all redshifts"],numpoints=1)
plt.savefig("colmag_millennium_z.png")
plt.show()
```
