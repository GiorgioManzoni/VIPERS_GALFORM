```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib 
matplotlib.rc('xtick', labelsize=10) 
matplotlib.rc('ytick', labelsize=10)
matplotlib.rcParams.update({'font.size': 20})
import pandas as pd
from virgodb import VirgoDB
vdb = VirgoDB("giorgio",password="BN401kmd")
import timeit
```

```python
#-- Select main progenitors of a particular galaxy 
Q = 'select des.mag_V_ext as d_mag_V_ext, des.SDSS_u_ext as d_SDSS_u_ext, prog.mag_V_ext as p_mag_V_ext, prog.SDSS_u_ext as p_SDSS_u_ext ,prog.stellarmass as p_stellarmass,des.stellarmass as d_stellarmass '#,prog.redshift as p_redshift, des.mdhalo as d_mdhalo, prog.mdhalo as p_mdhalo ' 
Q += 'from  Gonzalez2014a..mr7 as prog, Gonzalez2014a..mr7 as des '
Q += 'where  des.snapnum = 47 and ' #redshift=0.456577 
Q += 'prog.galaxyid between des.galaxyid and des.mainleafid '
Q += 'and prog.snapnum = 37 and des.random<0.0001 ' #redshift=1.173417
Q += 'and des.stellarmass>0.1'
prog_37_47 = vdb.execute_query(Q)
```

```python
len(prog_37_47['p_mag_V_ext'])
```

```python
plt.figure(figsize=(10,8))
plt.plot(prog_37_47['d_mag_V_ext'],prog_37_47['d_SDSS_u_ext']-prog_37_47['d_mag_V_ext'],'.')
plt.xlabel("mag_V_ext")
plt.ylabel("SDSS_U_ext - mag_V_ext")
plt.title('descendant (z=0.46)')
plt.xlim(-22.5,-16.5)
plt.ylim(0.3,2.25)
plt.savefig("../plots/007/col_mag_des.png")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(prog_37_47['p_mag_V_ext'],prog_37_47['p_SDSS_u_ext']-prog_37_47['p_mag_V_ext'],'.',c='C1')
plt.xlabel("mag_V_ext")
plt.ylabel("SDSS_U_ext - mag_V_ext")
plt.title('progenitor (z=1.17)')
plt.xlim(-22.5,-16.5)
plt.ylim(0.3,2.25)
plt.savefig("../plots/007/col_mag_prog.png")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.plot(prog_37_47['d_mag_V_ext'],prog_37_47['d_SDSS_u_ext']-prog_37_47['d_mag_V_ext'],'.',label=r'des $z\sim0.46$')
plt.plot(prog_37_47['p_mag_V_ext'],prog_37_47['p_SDSS_u_ext']-prog_37_47['p_mag_V_ext'],'.',label=r'pro $z\sim1.17$')
plt.xlabel("mag_V_ext")
plt.ylabel("SDSS_U_ext - mag_V_ext")
plt.legend()
plt.xlim(-22.5,-16.5)
plt.ylim(0.3,2.25)
plt.savefig("../plots/007/col_mag_des_prog.png")
plt.show()
```

```python
h = 'd_mag_V_ext \t d_SDSS_u_ext \t p_mag_V_ext \t p_SDSS_u_ext \t p_stellarmass \t d_stellarmass' #\t p_redshift \t d_mdhalo \t p_mdhalo '
np.savetxt("../data/progenitor_VIPERS_like_37_47.txt",prog_37_47.T, header=h, encoding='utf8')
```
