```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib 
matplotlib.rc('xtick', labelsize=10) 
matplotlib.rc('ytick', labelsize=10)
matplotlib.rcParams.update({'font.size': 20})
import pandas as pd
from fase.fileio import Table
from cosmology import *
from edge import *
import sys
```

```python
zspec = W1.getCol('zspec')
K_corr = W1.getCol('kcorr_i')
type_gal = W1.getCol('CE_PT')
magV = W1.getCol('M_V')
magU = W1.getCol('M_Uj')
u_T07 = W1.getCol('u_T07')
#v_T07 = W1.getCol('')
magI = W1.getCol('M_I')
i_CFHTLS = W1.getCol('M_i_CFHTLS')
DM = W1.getCol('DM')
i_T07 = W1.getCol('i_T07')
g_T07 = W1.getCol('g_T07')
```

```python
x =np.arange(0.,2.,0.001)
y1 = 1.2*x**2
y2 = 0.9*x**3
y3 = 0.5*x**3-0.1
y4 = -0.25*(0.0*x+1.)
```

```python
plt.figure(figsize=(10,8))

plt.plot(zspec[type_gal==1],K_corr[type_gal==1],".",markersize=1,label=r'type 1 (ellipticals) $y = 1.2 \,x^2$')
plt.plot(zspec[type_gal==2],K_corr[type_gal==2],".",markersize=1,label=r'type 2 (early spirals) $y = 0.9 \, x^3$')
plt.plot(zspec[type_gal==3],K_corr[type_gal==3],".",markersize=1,label=r'type 3 (spirals) $y = 0.5 \, x^3 -0.1$')
plt.plot(zspec[type_gal==4],K_corr[type_gal==4],".",markersize=1,label=r'type 4 (starbursts) $y = -0.25$')

plt.plot(x,y1,lw=3,color='blue')
plt.plot(x,y2,lw=3,color='orange')
plt.plot(x,y3,lw=3,color='green')
plt.plot(x,y4,lw=3,color='red')

plt.xlim(0.,1.5)
plt.ylim(-1.5,2.5)
plt.title('VIPERS galaxies in W1')
plt.xlabel('zspec')
plt.ylabel('k-correction i-band')
plt.legend(markerscale=10,fontsize='small')
plt.savefig('../plots/012/K-correction_VIPERS.png')
plt.show()
```

```python
sel_1 = np.where(type_gal==1)
sel_2 = np.where(type_gal==2)
sel_3 = np.where(type_gal==3)
sel_4 = np.where(type_gal==4)
sel_99 = np.where(type_gal==-99)
```

```python
plt.figure(figsize=(10,8))
plt.plot(magV[sel_1],magU[sel_1]-magV[sel_1],".",markersize=1,alpha=0.7,label='type 1')
plt.plot(magV[sel_2],magU[sel_2]-magV[sel_2],".",markersize=1,alpha=0.7,label='type 2')
plt.plot(magV[sel_3],magU[sel_3]-magV[sel_3],".",markersize=1,alpha=0.7,label='type 3')
plt.plot(magV[sel_4],magU[sel_4]-magV[sel_4],".",markersize=1,alpha=0.7,label='type 4')
plt.ylabel("(magU - magV) rest-frame")
plt.xlabel("mag V rest-frame (h=0.7)")
plt.xlim(-24,-18)
plt.ylim(0.,2.5)
plt.legend(markerscale=30)
plt.savefig('../plots/012/colmag_type.png')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(magU[sel_1]-magV[sel_1],bins=30,range=(0,2.5),label='type 1',histtype='step',lw=4)
plt.hist(magU[sel_2]-magV[sel_2],bins=30,range=(0,2.5),label='type 2',histtype='step',lw=4)
plt.hist(magU[sel_3]-magV[sel_3],bins=30,range=(0,2.5),label='type 3',histtype='step',lw=4)
plt.hist(magU[sel_4]-magV[sel_4],bins=30,range=(0,2.5),label='type 4',histtype='step',lw=4)
plt.axvline(np.median(magU[sel_1]-magV[sel_1]),color='C0',lw='3',ls='--',label="median = "+str(round(np.median(magU[sel_1]-magV[sel_1]),3)))
plt.axvline(np.median(magU[sel_2]-magV[sel_2]),color='C1',lw='3',ls='--',label="median = "+str(round(np.median(magU[sel_2]-magV[sel_2]),3)))
plt.axvline(np.median(magU[sel_3]-magV[sel_3]),color='C2',lw='3',ls='--',label="median = "+str(round(np.median(magU[sel_3]-magV[sel_3]),3)))
plt.axvline(np.median(magU[sel_4]-magV[sel_4]),color='C3',lw='3',ls='--',label="median = "+str(round(np.median(magU[sel_4]-magV[sel_4]),3)))

plt.xlabel("(magU - magV) rest-frame")
plt.ylabel("counts")
#plt.xlim(-24,-18)
#plt.ylim(0.,2.5)
plt.legend(markerscale=30,loc=2,fontsize='x-small')
plt.savefig('../plots/012/histo_col_type.png')
plt.show()
```
